```python
import cv2
import numpy as np
target_image = cv2.imread('1.jpg', cv2.IMREAD_GRAYSCALE)
scene_image = cv2.imread('1.jpg', cv2.IMREAD_GRAYSCALE)
orb = cv2.ORB_create()
keypoints_target, descriptors_target = orb.detectAndCompute(target_image, None)
keypoints_scene, descriptors_scene = orb.detectAndCompute(scene_image, None)
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(descriptors_target, descriptors_scene)
matches = sorted(matches, key=lambda x: x.distance)
N = 10
matching_result = cv2.drawMatches(target_image, keypoints_target,scene_image, keypoints_scene,
matches[:N], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
cv2.imshow('Feature Matching Result', matching_result)
import matplotlib.pyplot as plt
plt.imshow(matching_result)
plt.axis('off')
plt.show()
```
